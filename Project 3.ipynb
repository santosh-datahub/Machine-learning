## ONe hot encoding

```python
x_one_encoed =pd.get_dummies(x,columns=['model','transmission', 'fuelType'] ,drop_first=True ) 
x_one_encoed
```

```python
x_one_encode =x_one_encoed.astype(int)  
x_one_encode
```

## Label encodeing

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

# Make a copy of X to avoid modifying original
Xlabel = x.copy()
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    Xlabel[col] = le.fit_transform(Xlabel[col].astype(str))  # Convert to string for safety
    label_encoders[col] = le
```

## using i[onehot encoding value chnage -3 and 3]

```python
from sklearn.preprocessing import StandardScaler

numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()

x_one_encode[numerical_cols] = scaler.fit_transform(x_one_encode[numerical_cols])
x_one_encode
```

```python
from sklearn.model_selection import train_test_split
from sklearn. linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

## this test perform by ONE SHOT ENCODING

## ðŸ”¹ Definitions

1) X â†’ Features (independent variables)

1) These are the input columns (like year, mileage, engineSize, etc.)

 1) 1-Used to predict something.

y â†’ Target (dependent variable)

The column you want to predict (like price).

After splitting:

X_train â†’ Feature values used to train the model.

X_test â†’ Feature values used to test the model after training.

y_train â†’ Target values (labels) corresponding to X_train.

y_test â†’ Target values (labels) corresponding to X_test (used for evaluation).

```python
 X_train, X_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.33, random_state=42)
```

 ## X_train, X_test â†’ features

 ## y_train, y_test â†’ target (e.g., price)

```python
model= LinearRegression() 
model.fit(X_train ,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

## to find the r2

 ## Measures how much of the variation in the target variable (y) can be explained by the independent variables (X).

```python
r2=r2_score(y_test ,y_predict)
r2
```

## Variance tells you how much the values differ from the average.

 ## High variance â†’ values are very spread out.

 ## Low variance â†’ values are close to the mean.

## to find adjusted r2 

* When comparing two or more regression models with different numbers of predictors.
* To check if adding new features is actually improving the model or just artificially inflating RÂ².



Example
* Model 1: Predict house price with size â†’ RÂ² = 0.75, Adjusted RÂ² = 0.74
* Model 2: Predict house price with size + location + random_noise â†’ RÂ² = 0.76, Adjusted RÂ² = 0.73
* ðŸ‘‰ RÂ² went up (0.75 â†’ 0.76), but Adjusted RÂ² went down (0.74 â†’ 0.73).
* This tells us that random_noise did not really help.

```python
n = X_test.shape[0]   # number of samples (rows in X_test)
p = X_test.shape[1]   # number of predictors/features (columns in X_test)

adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted RÂ² Score:", adjusted_r2)
```
